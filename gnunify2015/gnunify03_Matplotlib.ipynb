# Matplotlib

Matplotlib is a 2D graph plotting library for Python that attempts to emulate graph plotting in Matlab(R). To plot graphs inside the notebook (inline), use the IPython magic command **`%matplotlib inline`**, otherwise graphs will be plotted in an independent window that pops up outside the browser. Matplotlib depends on NumPy for the array implementation. It is customary to import **`matplotlib.pyplot`** as **`plt`**

```python
from __future__ import print_function, division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0, 4*np.pi, 201)
y = np.sin(x)
plt.plot(x, y)
plt.axis(ymin=-1.1, ymax=1.1)
plt.grid()
plt.title('Graph of $cos x$')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.savefig('fig01.png')
```

```python
people = ('Raj', 'Vijay', 'Rahul', 'Ajay', 'Vivek')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people)) # Generate randomly
error = np.random.rand(len(people)) # Generate randomly

plt.barh(y_pos, performance, xerr=error, align='center', alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel(performance)
plt.title('How fast do you want to go today?')
```

### Multiple lines, legend, $\LaTeX{}$ typesetting

```python
x = np.linspace(0, 2*np.pi, 201)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.grid()
plt.legend((r'sin(x)', r'cos(x)'), shadow=True, loc='lower left')
plt.title('Harmonic Functions')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$f(\theta)$')
```

## Stem Plot

```python
xx = np.linspace(0, 10, 21)
yy = 50 * np.random.rand(len(xx))
plt.stem(xx, yy, markerfmt='co', linefmt='r-')
plt.xlim(min(xx)-1, max(xx)+1)
plt.ylim(0, max(yy)+2)
plt.show()
```

## Object Oriented Interface to Matplotlib

```python
x, y = np.random.randn(2, 100)

fig = plt.figure()

ax1 = fig.add_subplot(211)
ax1.xcorr(x, y)
ax1.grid(True)
ax1.axhline(0, color='black', lw=2)

ax2 = fig.add_subplot(212)
ax2.acorr(x)
ax2.grid(True)
ax2.axhline(0, color='black', lw=2)

plt.show()
```
