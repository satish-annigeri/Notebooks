# Display of data through Graphs

```python
%matplotlib inline
from __future__ import print_function, division
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#
df = pd.DataFrame(np.array([[1, 4.6, 12.1], [2, 11.6, 30.5], [3, 7.4, 19.5], [4, 3.3, 8.7], \
[5, 8.6, 22.6], [6, 2.5, 6.6]]), columns = ['category', 'count', 'percentage']);
print (df);
sns.set(style = 'whitegrid');
plt.autoscale (tight = 'x');
plt.tight_layout();
plt.ylim([0, 35]);
colors = ['#FF0000', '#FFFF00', '#00FF00', '#00FFFF', '#0000FF', '#FF00FF'];
plt.bar(df['category'], df['percentage'], width = 0.5, color = colors, \
edgecolor = '#000000', linewidth = 2);
```

```python
%matplotlib inline
import pandas as pd
df = pd.DataFrame(np.array([[1, 4.6, 12.1], [2, 11.6, 30.5], [3, 7.4, 19.5], [4, 3.3, 8.7], \
[5, 8.6, 22.6], [6, 2.5, 6.6]]), columns = ['category', 'count', 'percentage']);
sns.set(style = 'whitegrid');
plt.autoscale (tight = 'x');
plt.tight_layout();
explode = np.array([0, 0, 0, 0, 0.2, 0]);
colors = sns.dark_palette("skyblue", 8, reverse = True);
labels = ['Not HS Grad', 'HS Grad', 'Some College', 'Associate', 'Bachelors', 'Advanced'];
plt.pie(df['percentage'], explode = explode, colors = colors, labels = labels,\
autopct = '%2.0f', shadow = True);
plt.show();
```
