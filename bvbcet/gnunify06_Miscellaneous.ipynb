## Vectorize Calculations
NumPy functions are optimized to work with arrays and are much faster than performing the computations in loops. This can be demonstrated by calculating the $\sin$ of an array of numbers using any **math** function inside a loop and using the corresponding NumPy function.

```python
import math
import numpy as np

def f1(x):
    y = np.zeros(x.shape, dtype=float)
    for i in range(len(x)):
        y[i] = math.sin(x[i])
    return x

def f2(x):
    return np.sin(x)

x = np.linspace(0, 10, 1001)
%timeit f1(x)
%timeit f2(x)
```

## Numba

Numba provides a Just In Time compiler to generate optimized machine code to speed up computations in Python. It uses LLVM compiler infrastructure.

```python
import numpy as np
from numba import f8, jit

def sum1d(x):
    s = 0.0
    for i in range(x.shape[0]):
        s += x[i]
    return s

@jit(f8(f8[:]))
def fast_sum1d(x):
    s = 0.0
    for i in range(x.shape[0]):
        s += x[i]
    return s

x = np.linspace(0, 100, 1001)

%timeit sum1d(x)
%timeit fast_sum1d(x)
```

Let us compute the approximate value of $\pi$ using the following sum:

$$\pi \approx \sqrt{6 \sum_{k=1}^{n+1} \frac{1}{k^2}}$$

```python
def approx_pi(n=10000):
    s = 0
    for k in range(1, n+1):
        s += 1.0 / k**2
    return (6 * s) ** 0.5

%timeit approx_pi(1000)
%timeit approx_pi(10000)

print(approx_pi(1000))
print(approx_pi(10000))
```

```python
from numba.decorators import autojit

@autojit
def fast_approx_pi(n):
    s = 0
    for k in range(1, n+1):
        s += 1.0 / k**2
    return (6 * s) ** 0.5

n = 10000
%timeit approx_pi(n)
%timeit fast_approx_pi(n)
```

# Image Processing using scikit-image
scikit-image is an image proessing module for SciPy.While SciPy has some image processing functions, scikit-image greatly enhances image processing capabilities of Python.

```python
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import data, io, filter
%matplotlib inline

```

```python
image = data.coins() # or any NumPy array!
plt.subplot(211)
io.imshow(image)
plt.subplot(212)
edges = filter.sobel(image)
io.imshow(edges)
plt.show()
```
