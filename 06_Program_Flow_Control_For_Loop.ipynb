# Program Flow Control - For Loop

**Remember, indentation is important**. Statements belonging to a **`for loop`** must all indented by the same amount.

First, let us get to know **`range()`** and **`xrange()`** functions that generate sequence of numbers. In fact, the function **`range()`** creates and returns a list, while **`xrange()`** returns a generator. A generator is a kind of a template to generate a list at a later point of time.

You will observe that the ending value is not included in the list of values generated by **`range()`** (or to be generated by **`xrange()`**.

Both **`range()`** and **`xrange()`** generate only a list of integers, but cannot generate a list of real numbers.

```python
x = range(5); print x, type(x) # Starting value defaults to 0. Increment defaults to 1
y = range(1, 6); print y       # Increment defaults to 1
z = range(2, 12, 2); print z
x = xrange(0, 5); print x, type(x)
# Try this y = range(1.5, 5.5, 0.5)
```

```python
for i in range(5):
    print i
```

```python
for j in range(1, 6):
    print j
```

```python
for k in range(2, 12, 2):
    print k
```

The difference between using **`range()`** and **`xrange()`** for looping is that in the former case, a list is created and each item in the list is used sequentially as the value of the looping variable. In the latter case, **`xrange()`** being a generator, instead of generating a list, one value in the sequence is generated each time the loop is executed. While the difference is minimal for short loops, **`xrange()`** is more efficient when the loops are long. For example, **`for i in xrange(1, 100001):`** is more efficient than **`for i in range(1, 100001):`**. You can use **`xrange()`** in exactly the same way as **`range()`**.

```python
for i in xrange(5):
    print i
```

```python
for i in xrange(10, 0, -2):
    print i
```

Since  **for loop** iterates over the items of a list, you can use any list (or any container, for that matter) to iterate over its items. Here we create a list and loop over its items.

```python
a = [1, 10, -5, 1]
for i in a:
    print i
```

Here we create a dictionary and loop over its items. But remember that dictionary does not maintain any specific order. You may have to sort the items by their key in case you want a specific order.

```python
b = {'a': 1, 'b': 2, 'c': 99}
for j in sorted(b):
    print b[j]
```

```python
words = ['cat', 'window', 'demonstrate']
for w in words:
    print w, # Trailing comma prevents a newline
```

```python
s = 'armageddon'
for c in s:
    print 2*c, ' ',
```

Let us time a for loop using **`range()`** and **`xrange()`** and compare thier speeds.

```python
%timeit for i in range(10000): pass
%timeit for i in xrange(10000): pass
```

## While Loop

A **`while`** is a general form of a **`for`** loop, and requires the looping variable to be initialized before looping begins, write the condition for continuation of looping and alter the looping variable inside the loop.

```python
# Fibonacci series
a, b = 0, 1 # Tuple unpacking
while b < 100:
    print a,
    a, b = b, a+b # Tuple unpacking again!
```

## `break`,  `continue` and `pass`

You can interrupt a loop (either **`for`** or **`while`**) with the **`break`** keyword. This will interrupt the current loop and continue as if the looping completed normally.

On the otherhand, **`continue`** interrupts the next item in the loop, as if the current iteration was normally completed.

The **`pass`** statement does nothing.It can be used when a statement is required syntactically but the program requires no action. It can also be used to create minimal functions or classes.

```python
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print n, 'equals', x, '*', n/x
            break
    else:
        # Loop fell through without finding a factor
        print n, 'is a prime number'
```

```python
for n in range(2, 10):
    if n % 2 == 0:
        print 'Found an even number', n
        continue
    print 'Found a number', n
```

```python
def myfunc():
    pass

class myclass:
    pass
```

While looping through a sequence, the position index and corresponding value can be retrieved at the same time using the **`enumerate`** function:

```python
for i, v in enumerate(['tic', 'tac', 'toe']):
    print i, v
```

To loop over two or more sequences at the same time, the entries can be paired with the **`zip()`** function:

```python
questions = ['name', 'quest', 'favorite color']
answers = ['Lancelot', 'the holy grail', 'blue']
for q, a in zip(questions, answers):
    print 'What is your {0}? It is {1}.'.format(q, a)
```
