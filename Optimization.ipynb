```python
from __future__ import division
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(-1.5, 1.5, 51)
y = np.linspace(-1, 1.5, 51)

x, y = np.meshgrid(x, y)

z = 100 * (y - x**2)**2 + (1 - x)**2

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_wireframe(x, y, z, rstride=2, cstride=2, cmap=cm.coolwarm)
```
