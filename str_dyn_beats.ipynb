# Graphical Representation of Beats
A mass is subjected to two harmonic displacements given by $x_1 (t)=X \cos \omega t$ and $x_2 (t)=X \cos(\omega + \delta)t$ with $X = 1$cm and $\omega = 2$ radians/sec & $\delta = 1$ radians/sec. Plot the resulting motion of the mass using MATLAB & identify the beat frequency of the mass $x(t)$.

##Solution
$x(t) = x_1(t) + x_2(t)$

$x(t) = X \cos \omega t + X \cos ((\omega + \delta) t)$

$x(t) = X \left[ \cos \omega t + X \cos ((\omega + \delta) t) \right]$

$x(t) = X \left[ 2 \cos(\frac{\omega t + \omega t + \delta t}{2}) \cos(\frac{\omega t - \omega t - \delta t}{2}) \right]$

$x(t) = 2 X \left[ \cos(\omega t + \frac{\delta t}{2}) \cos(-\frac{\delta t}{2})  \right]$

$x(t) = 2 X \left[ \cos \left(\omega + \frac{\delta}{2} \right) t \cdot \cos \left(\frac{\delta}{2} \right) t \right]$

### Example 1
$X = 1$cm, $\omega = 20$rad/s and $\delta = 1$ rad/s

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Set the parameters
X=2.0        # Amplitude in cm
w = 20.0     # Frequency in rad/s
delta = 1.0  # Difference in frequency of second displacement with respect to the first, in rad/s

t = np.linspace(0, 15, 1001)      # Create 1001 equally spaced points starting from 0 up to 15
```

We can calculate $x(t)$ as the superposition of $x_1(t)$ and $x_2(t)$ and then plot $x(t)$.

```python
x1 = X * np.cos(w * t)            # Create x1(t)
x2 = X * np.cos((w + delta) * t)  # Create x2(t)
xx = x1 + x2
plt.plot(t, xx)
plt.grid()
plt.xlabel('$t$ (sec)')
plt.ylabel('X(t)')
plt.title('Beat Phenomenon $X=1, \omega=20, \delta=1$')
plt.show()
```

Alternately, we can calculate $x(t)$ from the expression derived above, namely

$x(t) =  2 X \left[ \cos \left(\omega + \frac{\delta}{2} \right) t \cdot \cos \left(\frac{\delta}{2} \right) t \right]$

```python
x = 4.0 * X * np.cos(delta * t / 2.0) * np.cos((w + delta / 2.0) * t)
plt.plot(t, x)
plt.grid()
plt.xlabel('$t$ (sec)')
plt.ylabel('X(t)')
plt.title('Beat Phenomenon $X=1, \omega=20, \delta=1$')
plt.show()
```

### Example 2
$X = 1$cm, $\omega = 20$rad/s and $\delta = 0$ rad/s

```python
delta = 0
x1 = X * np.cos(w * t)
x2 = X * np.cos((w + delta) * t)
xx = x1 + x2
plt.plot(t, xx)
plt.grid()
plt.xlabel('$t$ (sec)')
plt.ylabel('X(t)')
plt.title('Beat Phenomenon $X=1, \omega=20, \delta=0$')
plt.show()
```

### Example 3
$X = 1$cm, $\omega = 20$rad/s and $\delta = 2$ rad/s

```python
delta = 2
x1 = X * np.cos(w * t)
x2 = X * np.cos((w + delta) * t)
xx = x1 + x2
plt.plot(t, xx)
plt.grid()
plt.xlabel('$t$ (sec)')
plt.ylabel('X(t)')
plt.title('Beat Phenomenon $X=1, \omega=20, \delta=2$')
plt.show()
```

##Interpretation

It is evident from the graphs obtained that, as the value of delta ($\delta$) changes from $\delta = 1$ to $\delta = 0$, the phenomenon of beats does not occur and only the simple harmonic motion is observed and when the value of delta change  $\delta = 1$ to $\delta = 2$, the frequency of beat phenomena increases, that is, the beat pattern occurring between the time interval $0 \leq t \leq 15$ increases
