# Graphical Representation of Fourier Series
Plot the function $x(t) = A \frac{t}{\tau}$ where $0 \leq t \leq \tau$ and its Fourier series representation with four terms.

$\bar{x}(t) = \frac{A}{\pi} \lbrace \frac{\pi}{2} - \left( \sin \omega t + \frac{1}{2} \sin 2 \omega t + \frac{1}{3} \sin 3 \omega t  \right) \rbrace$ for $0 \leq t \leq \tau$ with $A = 1$, $omega = \pi$ and $\tau = \frac{2 \pi}{\omega} = 2$.

### One Term
$\bar{x_1}(t) = \frac{A}{2}$

### Two Terms
$\bar{x_2}(t) = \frac{A}{2} - \frac{A}{\pi} \sin \omega t$

### Three Terms
$\bar{x_3}(t) = \frac{A}{2} - \frac{A}{\pi} \sin \omega t - \frac{A}{2 \pi} \sin 2 \omega t$

### Four Terms
$\bar{x_3}(t) = \frac{A}{2} - \frac{A}{\pi} \sin \omega t - \frac{A}{2 \pi} \sin 2 \omega t - - \frac{A}{3 \pi} \sin 3 \omega t$

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

A = 1.0
pi = np.pi
w = pi
tau = 2.0

t = np.linspace(0, tau, 1001)
x0 = A * t / tau
x1 = np.ones(t.shape, dtype=float) * A / 2.0
x2 = x1 - A / pi * np.sin(w * t)
x3 = x2 - A / (2.0 * pi) * np.sin(2.0 * w * t)
x4 = x3 - A / (3.0 * pi) * np.sin(3.0 * w * t)

plt.subplot(231)
plt.plot(t, x0)
plt.grid()
plt.title('$x(t) = A \,t / \\tau} $')
plt.xlabel('t')

plt.subplot(232)
plt.plot(t, x1)
plt.grid()
plt.title('One Term')
plt.xlabel('t')

plt.subplot(233)
plt.plot(t, x2)
plt.grid()
plt.title('Two Terms')
plt.xlabel('t')

plt.subplot(234)
plt.plot(t, x3)
plt.grid()
plt.title('Three Terms')
plt.xlabel('t')

plt.subplot(235)
plt.plot(t, x4)
plt.grid()
plt.title('Four Terms')
plt.xlabel('t')

plt.tight_layout()
plt.show()
```

Replot the function with $\omega = 2 \pi$.

```python
A = 1.0
pi = np.pi
w = 2.0 * pi
tau = 2.0

t = np.linspace(0, tau, 1001)
x1 = np.ones(t.shape, dtype=float) * A / 2.0
x2 = x1 - A / pi * np.sin(w * t)
x3 = x2 - A / (2.0 * pi) * np.sin(2.0 * w * t)
x4 = x3 - A / (3.0 * pi) * np.sin(3.0 * w * t)

plt.subplot(221)
plt.plot(t, x1)
plt.grid()
plt.xlabel('t')

plt.subplot(222)
plt.plot(t, x2)
plt.grid()
plt.xlabel('t')

plt.subplot(223)
plt.plot(t, x3)
plt.grid()
plt.xlabel('t')

plt.subplot(224)
plt.plot(t, x4)
plt.grid()
plt.xlabel('t')

plt.show()
```

### Interpretation
It is evident from the plots that when frequency $\omega$ value is increased from $\pi$ to $2 \pi$, the number of cycles in Fourier series increases and when frequency $\omega$ is made zero no harmonic motion occurs.
